```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
def tp(A,B,C):
    return np.kron(np.kron(A,B),C)
state_0 = np.array([1,0])
state_1 = np.array([0,1])
proj_0 = np.outer(state_0,state_0) # top half of the identity matrix
proj_1 = np.outer(state_1,state_1) # bottom half of the identity matrix
X = np.array([[0,1],[1,0]])
U_cnot = np.kron(proj_0,np.eye(2))+np.kron(proj_1,X)

def f(i,j):
    a=state_0
    if i==0:
        b=state_0
    else:
        b=state_1
    if j==0:
        c=state_0
    else:
        c=state_1
    return tp(a,b,c)
def Ha(s):
    H = (1/np.sqrt(2))*np.array([[1,1],[1,-1]]) 
    U_g = np.kron(np.kron(H,np.eye(2)),np.eye(2))
    return U_g@s
def double_cnot(s):
    U=np.kron(np.kron(proj_0,np.eye(2)),np.eye(2))+np.kron(np.kron(proj_1,X),X)
    return U@s
def final(s):
    g=Ha(s)
    g1=double_cnot(g)
    g2=Ha(g1)
    return g2
s=f(0,0)

final(s)

    
```
